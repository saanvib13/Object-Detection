```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

#### setting up the model

```python
config_file='ssd_mobilenet_v3_large_coco_2020_01_14.pbtxt'
mobile_net_model='frozen_inference_graph.pb'
```

```python
model_setup=cv2.dnn.DetectionModel(mobile_net_model,config_file)
```

```python
model_setup.setInputSize(320,320)
model_setup.setInputScale(1.0/127.5)
model_setup.setInputMean((127.5,127.5,127.5))
model_setup.setInputSwapRB(True)
```

#### classes that can be detected

```python
classes=[]
with open('labels.txt','rt') as file:
    classes=file.read().split('\n')
```

```python
np.array(classes).shape
len(classes)
# print(classes) #90
classes
```
